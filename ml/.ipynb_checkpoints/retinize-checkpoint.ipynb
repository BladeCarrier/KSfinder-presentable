# Что тут происходит:
 * здесь предпосчитываются Retina-представления данных

```python
import os
import pandas as pd
from sklearn.externals import joblib
import random
from retina.retina_theanized import Retina
import numpy as np
```

```python

def get_retina_response(hits,retina):


    response = retina.response(hits,block_size=None,
                n_jobs=1,inner_block_size=None)

    return response


def retinize(folder,out_file,retina,max_rows = float('inf'),n_closest=5000):
    responses = []
    fpaths = map(lambda line:os.path.join(folder,line) , os.listdir(folder))
    i = 0
    for fpath in fpaths:

        hits = pd.DataFrame.from_csv(fpath)[["X","Y","Z"]].values
        #filter hits
        hits = hits[hits[:,2] > 100]
        
        hits_dist = np.linalg.norm(hits-retina.ks,axis=-1)
        
        hits = hits[np.argsort(hits_dist)[:n_closest],:]
        
        
        if i%100 ==0:
            print i,'/',min(len(fpaths),max_rows)
        
        if len(hits)==0:
            continue
        responses.append(joblib.delayed(get_retina_response)(hits,retina))
        
        i +=1
        if i >=max_rows:
            break

    responses = joblib.Parallel(n_jobs = -1)(responses)
    retina_pts = np.vstack(responses)
    np.save(out_file,retina_pts)
```

# Retina, которая за velo

```python
ks = np.zeros(3)
ks[2]+=2000
xdim = ydim = 64
dalpha = np.pi/8
dbeta = np.pi/8
alpha = 0
beta = 0
alphas = np.linspace(-dalpha+alpha, alpha+dalpha,num=xdim)
bethas = np.linspace(-dbeta+beta, beta+dbeta,num=ydim)
sigma = 50
#ks = KS[evtids_ks[i]][2]

retina = Retina(ks,alphas,bethas,variance=sigma,power=2)
```

```python
retinize("/mnt/KSfinder/mc_bg_down_hits/","/mnt/KSfinder/mc_bg_down_retina0.npy",retina,5000)
retinize("/mnt/KSfinder/mc_bg_up_hits/","/mnt/KSfinder/mc_bg_up_retina0.npy",retina,5000)
retinize("/mnt/KSfinder/mc_sig_hits/","/mnt/KSfinder/mc_sig_retina0.npy",retina,5000)
```
