```python
ks = np.zeros(3)
#ks[2]+=1
xdim = ydim = 64
ks[0] = 625.645
ks[1] = -1621.71
ks[2] = 10745
dalpha = np.pi/3
dbeta = np.pi/3
alpha = -0.319414
beta = 0.708715
alphas = np.linspace(-dalpha+alpha, alpha+dalpha,num=xdim)
bethas = np.linspace(-dbeta+beta, beta+dbeta,num=ydim)
sigma = 50
#ks = KS[evtids_ks[i]][2]
```

```python

retina = Retina(ks,alphas,bethas,variance=sigma,power=2)
```

```python

def get_retina_response(hits):


    response = retina.response(hits,block_size=None,
                n_jobs=1,inner_block_size=None)

    return response

def retinize(folder,out_file,retina):
    responses = []
    fpaths = map(lambda line:os.path.join(folder,line) , os.listdir(folder))
    for i in range(len(fpaths)):
        fpath = fpaths[i]

        hits = pd.DataFrame.from_csv(fpath)[["X","Y","Z"]].values
        #hits = hits[hits[:,2] <= 5000]
        hits = hits[hits[:,2] > 100]
        if i%25 ==0:
            print i,'/',len(fpaths)
        if len(hits)==0:
            continue


        responses.append(joblib.delayed(get_retina_response)(hits))
    responses = joblib.Parallel(n_jobs = -1)(responses)
    retina_pts = np.vstack(responses)
    np.save(out_file,retina_pts)
```

```python
retinize("../eventtracer/mc_bg_hits/","./mc_bg_retina.npy",retina)
retinize("../eventtracer/mc_sig_hits/","./mc_ks_retina.npy",retina)
```
