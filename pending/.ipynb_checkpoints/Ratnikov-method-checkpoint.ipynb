```python
import numpy as np
import pandas as pd
```

```python
xyz = pd.DataFrame.from_csv("/mnt/KSfinder/mc_sig_hits/03695402_0000106558.hits.csv")[["X","Y","Z"]].values
```

# Algorithm
 * для каждой пары точек с различным Z проводим трёхмерную прямую
     * на самом деле для всех точек, которые отстоят по Z хотя бы на константу(100)
 * для каждой такой прямой находим точку, ближайшую к оси Z
     * а потом смотрим на все эти точки

```python
#take all pairs of points that are farther than 100 from one another along Z axis
z_tolerance = 100
pairs = np.array([(p0,p1) for i,p0 in enumerate(xyz) for p1 in xyz[i+1:] if abs(p1[2]-p0[2]) > z_tolerance]).astype('float32')
pairs.shape
```

```python
#get unit direction vectors for each line
dir_vectors = np.diff(pairs,axis = 1)
dir_vectors = dir_vectors.reshape(len(dir_vectors),-1)
dir_vectors /= np.linalg.norm(dir_vectors,ord=2,axis=-1,keepdims=True)
dir_vectors.shape
```

```python
#solve for a point closest to z in parametric system
# x = x0 +t*dir_vector

x0 = pairs[:,0,0]
y0 = pairs[:,0,1]
vx = dir_vectors[:,0]
vy = dir_vectors[:,1]

#optimal parameter
t = -(x0*vx + y0*vy) / (vx*vx + vy*vy)

#fix for cases where dir_vector is parallel to Z
t[np.isnan(t)] = 0

closest = pairs[:,0] + t[:,np.newaxis]*dir_vectors 
closest.shape
```

# sanity check
Проверка выкладок на вшивость

```python
distance_to_z = lambda pts:  np.linalg.norm(pts[:,:2],axis = -1)
```

```python
float_window = 0.001
step = 10

dist_at = distance_to_z(closest)
dist_after = distance_to_z(closest+dir_vectors*step)
dist_before = distance_to_z(closest-dir_vectors*step)
```

```python
print dist_after[:5]
print dist_at[:5]
print dist_before[:5]
```

```python
print np.all( dist_at <= (dist_after + float_window) )
print np.all( dist_at <= (dist_before + float_window) )
```
